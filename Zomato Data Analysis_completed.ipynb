```python
df_country = pd.read_excel('country-code.xlsx')
```

```python
df_country.head()
```

```python
df.columns
```

```python
#Merging country code with additional available file(county code) - see in the extreme left 
#pd.merge(df,df_country,on ='Country Code', how ='left'(Left Join))


final_df = pd.merge(df,df_country,on ='Country Code', how ='left')
```

```python
final_df.head()
```

```python
## To check Data Types

final_df.dtypes
```

```python
final_df.columns
```

```python
#Country Column count

final_df.Country.value_counts()
```

```python
# All the country names to plot pie chart
#final_df.Country.value_counts().index

Country_names=final_df.Country.value_counts().index
Country_names
```

```python
#Value of all the country to plot pie chart
#final_df.Country.value_counts().values

country_val = final_df.Country.value_counts().values
country_val
```

```python
# pie chart
#plt.pie(country_val,labels = Country_names)

# pie chart- Top 3 countries that uses Zomato with %
plt.pie(country_val[:3],labels = Country_names[:3],autopct="%1.2f%%")
```

Observation - Zomato maximum records or transaction are from india after that USA and United kingdoms

```python
final_df.columns
```

```python
# Grouping 3 columns first -> size().reset_index()- all data with indexing(reset) and column(0)
#final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()

# Rename the 0 column with Rating count
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings
```

# Conclusions - 
1. when the rating is between 4.5 to 4.9--> Excellent
2. when the ratings are between 4.0 to 4.4--> Very Good
3. when the ratings are between 3.5 to 3.9-->  Good
4. when the ratings are between 2.5 to 3.4-->  Average 
5. when the ratings are between 1.8 to 2.4--> Poor

```python
ratings.head()
```

```python
# to increase the size of figure
plt.rcParams['figure.figsize'] = (16,9) 

#To create barplot
sns.barplot(x= 'Aggregate rating', y='Rating Count', data = ratings)
```

```python
# Mapping the colors respect to 'rating color' 

sns.barplot(x= 'Aggregate rating', y='Rating Count',hue= 'Rating color', data = ratings, palette=['blue','red','orange','Yellow','Green','Green'])
```

Obseration:
1. No rated count is very high 
2. Maximum number of rating are between 2.5  to 3.4 

```python
## Count plot ->  frequency of records like No rating(blue) has only one record, Average(orange) rating has 10 records.

sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
